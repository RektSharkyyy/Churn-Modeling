### 2. Basic Proccessing

```python
df = pd.read_csv('data/processed/ChurnModeling_Outliers_Handled.csv')
df.head()
```

```python
df['Age'].min(), df['Age'].max()
```

```python
def custom_binning_credit_score(score):
    if score < 580:
        return 'Poor'
    if score < 670:
        return 'Fair'
    if score < 740:
        return 'Good'
    if score < 800:
        return 'Very Good'
    if score <= 850:
        return 'Excellent'
    else:
        assert True, "Credit score Cant go beyond 850"

df['CreditScoreBins'] = df['CreditScore'].apply(custom_binning_credit_score)
del df['CreditScore']

df.head(10 )
```

```python
df.to_csv('data/processed/ChurnModeling_Binning_Applied.csv', index=False)
```
