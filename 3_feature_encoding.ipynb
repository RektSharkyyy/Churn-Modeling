```python
df = pd.read_csv('data/processed/ChurnModeling_Binning_Applied.csv')
df.head(10)
```

### 2. Encode Nominal Variables

```python
norminal_variables = ['Geography', 'Gender']

geography_dummies = pd.get_dummies(df['Geography'], prefix="Geography")
gender_dummies = pd.get_dummies(df['Gender'], prefix="Gender")

df_encoded = pd.concat([df, geography_dummies], axis=1)
del df_encoded['Geography']

df_encoded = pd.concat([df_encoded, gender_dummies], axis=1)
del df_encoded['Gender']

df_encoded
```

### 2. Encode Ordinal Variables

```python
encode_dict_creditscore = {
                            'Poor' : 0,
                            'Fair' : 1,
                            'Good' : 2,
                            'Very Good' : 3,
                            'Excellent' : 4    
                            }
df['CreditScoreBins'] = df_encoded['CreditScoreBins'].map(encode_dict_creditscore)
df.head(10)
```

```python
df.to_csv('data/processed/ChurnModeling_Encoded.csv', index=False)
```
